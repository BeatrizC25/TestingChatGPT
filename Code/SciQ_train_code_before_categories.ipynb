## Re-Structure Dataset

```python
#delete support collumn cuz it wont be relevant in the future

df_SciQ=df_SciQ.drop('support', axis=1)
df_SciQ #confirm if it was done
```

```python
#check duplicates 
df_SciQ.duplicated(df_SciQ.columns).sum()
```

```python
#check for missing values
df_SciQ.isna().sum()
```

```python
#index starts with 0. we want it to start with 1
df_SciQ=df_SciQ.reset_index(drop=True)
```

```python
df_SciQ
```

```python
#create new dataframe with only question and correct answer and options
columns_to_copy=['question', 'correct_answer', 'distractor1', 'distractor2', 'distractor3']
df_copy = df_SciQ[columns_to_copy].copy()
```

```python
df_copy
```

```python
df_copy['random_letter'] = np.random.choice(['A', 'B', 'C', 'D'], size=len(df_copy))
```

```python
df_copy
```

```python
df_copy['A']='nothing'
```

```python
df_copy['B']='nothing'
df_copy['C']='nothing'
df_copy['D']='nothing'
```

```python
df_copy
```

```python
df_test=df_copy.copy()

for index, row in df_test.iterrows():
    correct_answer = row['correct_answer']
    random_letter = row['random_letter']
    
    # Fill in the correct answer in the corresponding column
    df_test.at[index, random_letter] = correct_answer
```

```python
df_test.head(3)
```

```python
for i in range(1, df_test.shape[0]+1):
    
    row_to_df=df_test.loc[[i]]
    
    options=['distractor1','distractor2', 'distractor3']

    if (row_to_df['A']=='nothing').bool():
        rand_int_A=np.random.randint(len(options)-1)
        new_A=df_test.loc[i, options[rand_int_A]]
        df_test.loc[i,'A']=new_A
        options.pop(rand_int_A)
    else: 
        pass
        
    if (row_to_df['B']=='nothing').bool():
        rand_int_B=np.random.randint(len(options)-1)
        new_B=df_test.loc[i, options[rand_int_B]]
        df_test.loc[i, 'B']=new_B
        options.pop(rand_int_B)
    else: 
        pass

    if (row_to_df['C']=='nothing').bool():
        #print('entrou no C')
        rand_int_C=np.random.randint(len(options) - 1) if len(options)>1 else 0
        new_C=df_test.loc[i, options[rand_int_C]]
        df_test.loc[i, 'C']=new_C
        options.pop(rand_int_C)
    else:
        pass
        
    if (row_to_df['D']=='nothing').bool(): 
        rand_int_D=np.random.randint(len(options) - 1) if len(options)>1 else 0
        new_D=df_test.loc[i, options[rand_int_D]]
        df_test.loc[i, 'D']=new_D
        
    else: 
        pass
        
```
