## Imports and Dataset Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore")
```

```python
json_file_path_1 = r"C:\Users\Bia\OneDrive\Ambiente de Trabalho\TESE_2024\DATASETS TESE\MedCQA\data\train.json"

# List to store JSON objects
data = []

with open(json_file_path_1, 'r') as json_file_1:
    for line in json_file_1:
        data.append(json.loads(line))

# Create a DataFrame from the loaded data
df_MedQA = pd.DataFrame(data)
```

## Re-Structure Dataset

```python
#check duplicated questions and nulls
df_MedQA.duplicated(subset=['question']).sum(), df_MedQA.isna().sum() 
```

```python
#nulls in exp do not matter because that column will be removed
df_MedQA.drop(columns=['exp'], inplace=True)
#id is also not important
df_MedQA.drop(columns=['id'], inplace=True)
```

```python
#we just want question with 1 answer. we will remove all rows with multi type of answer
df_MedQA = df_MedQA[df_MedQA['choice_type'] != 'multi']
```

```python
df_MedQA.head()
```

```python
df_MedQA.reset_index(drop=True)
```

```python
#delete choice type column
df_MedQA.drop(columns=['choice_type'], inplace=True)
```

```python
#change columns names
df_MedQA.rename(columns={'opa': 'A', 'opb': 'B', 'opc': 'C', 'opd': 'D'}, inplace=True)
```

```python
#create column Correct Option
df_MedQA[['Correct Option']]=None 
```

```python
#populate the new column
#first: create a list with options
options=['A', 'B', 'C', 'D']

#second: note how the number the column 'cop' is the index of the correct option in options minus 1 
#this means that number 1 is equivelent to A, 2 to B, etc

#third: assign the correct letter 
for index, row in df_MedQA.iterrows():
    ind=df_MedQA.loc[index, 'cop']-1
    df_MedQA.loc[index, 'Correct Option']=options[ind]
```

```python
#check results
df_MedQA.head()
```

```python
#it worked so we can remove 'cop'
df_MedQA.drop(columns=['cop'], inplace=True)
```

```python
#delete rows with unknown subject_name
df_MedQA = df_MedQA[df_MedQA['subject_name'] != 'Unknown']
```

```python
#delete rows with empty topic_name
df_MedQA = df_MedQA.dropna(subset=['topic_name'])
```

```python
#too many unique values, drop column
df_MedQA.drop('topic_name', axis=1, inplace=True)
```

```python
#rename columns 
df_MedQA.rename(columns={'question': 'Question', 'subject_name': 'Subject'}, inplace=True)
```

```python
#reorder columns
new_order=['Question', 'A', 'B', 'C', 'D', 'Correct Option', 'Subject']
df_MedQA=df_MedQA[new_order]
```

```python
#reset index and add 1
df_MedQA=df_MedQA.reset_index(drop=True)
```
