## Re-Structure Dataset

```python
#check for duplicates 
df_MathQA.duplicated(subset=['Problem']).sum()
```

```python
df_MathQA.head()
```

```python
#remove columns that will not be needed in the future: formulas and rationale

df_MathQA.drop(columns=['annotated_formula','linear_formula', 'Rationale'], inplace=True)
```

```python
#check
df_MathQA.head()
```

```python
#check for nulls
df_MathQA.isna().sum()
```

```python
#get value counts of category
df_MathQA['category'].value_counts()
```

```python
#change columns names and make the first letter of each category uppercase
df_MathQA['category']=df_MathQA['category'].str.title()
df_MathQA.rename(columns={'category': 'Sub-Branch'}, inplace=True)
df_MathQA.rename(columns={'Problem': 'Question'}, inplace=True)
```

```python
#check changes
df_MathQA.head()
```

```python
# Extract values associated with each letter from options into separate columns
df_math_test=df_MathQA.copy()
```

```python
columns = ['A', 'B', 'C', 'D', 'E']
for col in columns:
    df_math_test[col] = None
```

```python
df_math_test.head()
```

```python
# Split the values and assign them to separate columns
for index, row in df_math_test.iterrows():
    values = row['options'].split(', ')
    #print (values)
    for i, value in zip(columns, values):
        value=str(value)
        indd=value.find(')')
        df_math_test.loc[index, i]= value[indd+1:]
    #for value in values:
        #print(value)
        #key, val = value[4:]
        #df_math_test.loc[index, key] = val
```

```python
#checking and praying 
df_math_test.head()
```

```python
#check for nulls
df_math_test.isna().sum()
```

```python
#since everything worked, lets make a final df
MathQA_train=df_math_test.copy()
```

```python
MathQA_train.head()
```

```python
#we can delete the esting df
del df_math_test
```

```python
MathQA_train.isna().sum()
```

```python
#delete options column
MathQA_train.drop(columns=['options'], inplace=True)
```

```python
#check
MathQA_train.head()
```

```python
# Convert all letters in the 'correct' column to uppercase
MathQA_train['correct'] = MathQA_train['correct'].apply(lambda x: ''.join(letter.upper() for letter in x))
```

```python
#check agaiin
MathQA_train.head()
```

```python
#change correct column name 
MathQA_train.rename(columns={'correct': 'Correct Option'}, inplace=True)
```

```python
#re order columns
new_order=['Question', 'A', 'B', 'C', 'D', 'E', 'Correct Option', 'Sub-Branch']
MathQA_train=MathQA_train[new_order]
```

```python
#final check
MathQA_train.head()
```

```python
#reset index
MathQA_train=MathQA_train.reset_index()
#add 1 to index
MathQA_train.index=MathQA_train.index+1
```
