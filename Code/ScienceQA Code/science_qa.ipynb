```python
df_test=new_df.copy()
```

```python
columns=['A', 'B','C','D']
for col in columns:
    df_test[col] = None
```

```python
df_test=df_test.reset_index(drop=True).copy() #just cuz i dont like strange indexes
```

```python
df_test.head()
```

```python
#assign chocies to the columns A, B, ect
for index, row in df_test.iterrows():
    #print(index)
    choices=row['choices']
    #print(choices[0])
    for i in range(len(choices)):
        #print(choices[i], columns[i])
        df_test.iloc[index][columns[i]]=choices[i]
```

```python
df_test.head()
```

```python
#adress correct option to column
#create column
df_test['Correct Option']=None
```

```python
#assign values
for index, row in df_test.iterrows():
    indd=df_test.iloc[index]['answer']
    df_test.iloc[index]['Correct Option']=columns[indd]
```

```python
df_test.head()
```

```python
#it worked so we will drop the columns with no interest
df_test.drop(columns=['choices', 'answer', 'task'], inplace=True)
```

```python
#change names of columns
df_test.rename(columns={'subject': 'Subject', 'topic': 'Topic', 'category': 'Category'}, inplace=True)
```

```python
#make first letter of values in those 3 columns upper case
df_test['Subject'] = df_test['Subject'].str.capitalize()
df_test['Topic'] = df_test['Topic'].str.capitalize()
df_test['Category'] = df_test['Category'].str.capitalize()
```
